```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
csv_file = 'https://gist.githubusercontent.com/danielmoralesp/650bf482113d05e75f283607637ea11b/raw/71b4ebd2472767662ffbd23fa35ec7ad6f5f92a6/heights.csv'
```

```python
data = pd.read_csv(csv_file)
```

```python
data.describe()
```

```python
plt.plot(data['height'], data['weight'])
plt.axis('tight')
```

```python
plt.plot(data['height'], data['weight'],
        marker='p',
        markerfacecolor='white',
        markeredgecolor='green',
        markeredgewidth='5',
        markersize='15',
        color='red',
        linestyle='None' 
    )
```

```python
plt.scatter(data['height'], data['weight'])
```

```python
plt.hexbin(data['height'], data['weight'])
```
