```python
import numpy as np
np.random.seed(10)

x1 = np.random.randint(10, size=6)
x2 = np.random.randint(10, size=(3, 4))
x3 = np.random.randint(10, size=(3, 4, 5))
```

```python
print(x1)
print(x2)
print(x3)
```

```python
print(x3.ndim) # number of dimensions ndim
print(x3.shape) # shape of the array
print(x3.dtype) # data type of the array
print(x3.size) # number of the elements in the array
print(x3.itemsize) # size of each element in the array
print(x3.nbytes) # total size of the array in bytes, mean size of each element * total number of elements
```
