groupby + rolling

```python
import pandas as pd

df = pd.DataFrame({
    "stock": ["TCS", "TCS", "TCS", "INFY", "INFY", "INFY"],
    "day": [1, 2, 3, 1, 2, 3],
    "price": [100, 110, 120, 200, 190, 210]
})
```

```python
df.groupby('stock')['price'].rolling(2).mean().reset_index(level=0, drop=False)
```

```python
df.groupby('stock')['price'].rolling(2).std().reset_index(level=0, drop=False)
```

```python
df['price'].rolling(3).mean()
```

```python
df['price'].rolling(3, min_periods=1).mean()
```

```python
df['price'].rolling(3, min_periods=1).std()
```

rolling().max(): rolling peak, e.g., 7‑day max temperature or 20‑day high in stock prices.

rolling().min(): rolling trough, e.g., lowest latency in last N samples or 20‑day low in prices.
