
# Groupby

```python
import pandas as pd

df = pd.DataFrame({
    "sector": ["IT", "IT", "Energy", "Energy", "IT"],
    "stock": ["INFY", "TCS", "RELIANCE", "ONGC", "WIPRO"],
    "price": [1000, 2000, 1500, 1200, 800],
    "qty": [10, 20, 15, 12, 8]
})
```

```python
df
```

```python
temp = df.groupby("sector")
temp["price"].mean()
```

```python
temp.mean()
```

```python
temp
```

```python
temp['price'].min()
temp['price'].max()
```

```python
temp.agg({
    'price': 'mean',
    'qty': 'sum'
})
```

```python
temp['price'].agg(['mean', 'min', 'max'])
```

```python
temp['qty'].agg(['min', 'max', 'sum', 'mean'])
```

```python
temp['price'].mean()
```

```python
temp['price'].mean().reset_index()
```

```python
df
```

```python
df.groupby('sector')[['price']].transform('mean').reset_index()
```

```python
df['price_to_sector'] = df['price'] / df.groupby('sector')['price'].transform('mean')
```
