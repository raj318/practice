```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 10, 100)

print(x)
fig, ax = plt.subplots(1, 2) # i think one axis and two figures? or two figures with each one axis

ax[0].plot(x, np.sin(x))
ax[0].set_title("sin graph")

ax[1].plot(x, np.cos(x))
ax[1].set_title('cos graph')

plt.show()
```

```python
fig, ax = plt.subplots(2, 2)

ax[0, 0].plot(x, x, label='plain')
ax[0, 0].legend()
ax[0, 1].plot(x, x**2, label='square')
ax[0, 1].legend()
ax[1, 0].plot(x, x**4, label='pow 4')
ax[1, 0].legend()
ax[1, 1].plot(x, np.sqrt(x), label='square root')
ax[1, 1].legend()


plt.show()

```

```python
plt.clf()
fig, ax = plt.subplots(2, 1, sharex=True)

ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x))
# fig.tight_layout()
fig.subplots_adjust(hspace=0.1, wspace=10)

plt.show()
```

```python
fix, ax = plt.subplots()
ax.plot(x, x**2, label='square')
ax.set_ylabel("sqare graph")

ax2 = ax.twinx()
ax2.plot(x, x**4)
ax2.set_ylabel("pow 4 graph")



plt.show()
```

```python
fix, ax = plt.subplots()
ax.plot(x**2, x, label='square')
ax.set_xlabel("sqare graph")

ax2 = ax.twiny()
ax2.plot(x**4, x)
ax2.set_xlabel("pow 4 graph")
```

```python
fig, ax = plt.subplots(figsize=(10, 10), dpi=300)
ax.plot(x, x**2, label='x square')
ax.set_title("dpi plots")
ax.legend()
ax.tick_params(labelsize=12)
plt.title("dpi testing graph")
plt.savefig('plot.png', bbox_inches='tight', dpi=300)
```

```python
fig, ax = plt.subplots(2, 2)

ax[0, 0].plot(x, x, label='plain')
ax[0, 0].legend()
ax[0, 0].set_title("plain graph")
ax[0, 1].plot(x, x**2, label='square')
ax[0, 1].legend()
ax[0, 1].set_title("square graph")
ax[1, 0].plot(x, x**4, label='pow 4')
ax[1, 0].legend()
ax[1, 0].set_title("pow 4 graph")
ax[1, 1].plot(x, np.sqrt(x), label='square root')
ax[1, 1].legend()

fig.suptitle("testing grid graphs", y=0.98)
plt.tight_layout(rect=[0, 0, 1, 0.99])
plt.show()
```
