```python
plt.style.use("seaborn-v0_8-whitegrid")
tdata = netflix_data['type'].value_counts()
gp = plt.bar(tdata.index, height=tdata.values, width=0.5)

# print(plt.xticks())
patches = gp.patches
for patch in patches:
    # t = [print(item) for item in dir(patch)]
    # break
    h = patch.get_height()
    w = patch.get_width()
    x = patch.get_x()
    y = patch.get_y()
    if x < 0.5:
        patch.set_x(x + w/4)
        patch.set_label(x + w/4)
        patch.set_color('blue')
        x_tick = 0 + 0.1
        plt.text(
            x= x + 0.4,
            y = h+1,
            s = str(h),
            ha = 'center',
            fontsize = 10,
            color = 'black',
            rotation = 45
        )
    else:
        patch.set_x(x - w/4)
        patch.set_label(x - w/4)
        patch.set_color('red')
        x1_tick = 1-0.1
        
        plt.text(
            x=x+0.13,
            y = h+0.5,
            s = str(h),
            ha = 'center',
            fontsize = 10,
            color = 'black',
            rotation = 45
        )
    
plt.xticks(ticks=[x_tick, x1_tick])
plt.xlabel("Type")
plt.ylabel("Count")
plt.legend(labels=["Movie", "TV Show"])
plt.title("Netflix Type graph")

plt.show()
```

top 5 countries, which has seen maximum number of movies between 2015 and 2019 year

```python
plt.figure(figsize=(10, 5))
netflix_data['release_year'] = netflix_data['release_year'].astype(int)
# netflix_data['release_year']
tdata = netflix_data[netflix_data['release_year'].isin(range(2015, 2019))]


temp = tdata['country'].value_counts().head(5)
plt.bar(temp.index, height=temp.values, width= 0.5)
plt.xlabel('Country')
plt.ylabel('Count')
plt.title('Netflix Movies by Country')

for i in range(len(temp)):
    plt.text(x=temp.index[i], y=temp.values[i]+50, s=str(temp.values[i]), ha='center', fontsize=10, color='black', rotation=45)

# plt.legend()
plt.show()
# tdata = tdata[tdata['release_year'].isin(range(2015, 2019))]
# tdata
```
