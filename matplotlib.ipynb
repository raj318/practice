```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

plt.plot(x, y) 
# or 
plt.plot([1, 2, 3, 4, 5], [10, 20, 25, 30, 35], color = 'orange')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Simple Line Plot')
# plt.show() # still works only due to the fact that we are in a ipynb file environement
```

```python
x = [1, 2, 3, 4]
y1 = [2, 4, 6, 8]
y2 = [1, 4, 9, 16]
y3 = [1, 7, 10, 16]

plt.plot(x, y1, label='Linear')
plt.plot(x, y2, label='Square')
plt.plot(x, y3, label='Y3')
plt.xlabel("X axis")
plt.ylabel("y1, y2, y3 axis")
plt.legend()
plt.title("Multiple line graphs")
plt.show()
```

```python
plt.plot(x, y1, color='blue', label='Linear', linestyle='-', marker='o')
plt.plot(x, y2, color='red', label='Square', linestyle='--', marker='X')
plt.legend()
plt.show()
```

```python
categories = ['A', 'B', 'C', 'D']
values = [10, 20, 30, 40]

plt.bar(categories, values, label='bar chart')
plt.xlabel('categories')
plt.ylabel('values')
plt.title('Basic bar chart')
plt.legend()
plt.show()
```

 Line plots


 Scatter plots


 Bar charts

These all require 1-to-1 pairing.

 If a plot connects dots  lengths must match


 If a plot summarizes data  one list is enough

```python
plt.barh(categories, values, label='horizontal bar chart', color='red')
plt.title("horizontal bar chart")
plt.legend()
plt.show()
```

```python
import numpy as np
print(categories)
x = np.arange(len(categories))
print(x)
width = 0.4
temp1 = x - width/2
print(temp1)
temp2 = x + width/2
print(temp2)

twidth = 0.3

plt.bar(temp1, values, width=twidth, label='bar 1', color='green')
plt.bar(temp2, values, width=twidth, label='bar 2', color='red')

plt.xticks(x, categories) # x is used to plot graphs and x is generated based on categories
plt.legend()
plt.show()
```

```python
bars = plt.bar(categories, values, label='line plot')
for bar in bars:
    plt.text(
        # bar.get_x() + bar.get_width()/2,
        bar.get_x() + bar.get_width()/2,
        bar.get_height(),
        str(bar.get_height()),
        ha='center', # what is this space in which left and right is setting?
        va='bottom'
    )
plt.show()
```

```python
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]
plt.scatter(x, y)

for xi, yi in zip(x, y):
    plt.text(xi+0.1, yi+0.1, str(yi), ha='center', va='bottom')

plt.show()
```
