```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

plt.plot(x, y) 
# or 
plt.plot([1, 2, 3, 4, 5], [10, 20, 25, 30, 35], color = 'orange')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Simple Line Plot')
# plt.show()
```

```python
x = [1, 2, 3, 4]
y1 = [2, 4, 6, 8]
y2 = [1, 4, 9, 16]
y3 = [1, 7, 10, 16]

plt.plot(x, y1, label='Linear')
plt.plot(x, y2, label='Square')
plt.plot(x, y3, label='Y3')
plt.xlabel("X axis")
plt.ylabel("y1, y2, y3 axis")
plt.legend()
plt.title("Multiple line graphs")
plt.show()
```

```python
plt.plot(x, y1, color='blue', label='Linear', linestyle='-', marker='o')
plt.plot(x, y2, color='red', label='Square', linestyle='--', marker='s')
plt.legend()
plt.show()
```

```python
plt.plot([1, 2, 3], [1, 2, 3])
plt.show()
```

```python
plt.plot([1, 2, 3,], [3, 2, 1])
plt.clf()
plt.plot([1, 2, 3,], [3, 2, 1])
plt.show()
```

```python
x = [1, 2, 3, 4]
y = [10, 5, 20, 8]

plt.plot(x, y)
plt.title("Line Plot")
plt.show()
```

```python
plt.scatter(x, y, color='red')
plt.title('Scatter Plot')
plt.show()
```

```python
sizes = [500, 100, 200, 80]
plt.scatter(x, y, s=sizes, color='red')
plt.show()
```

```python
plt.scatter(x, y, alpha=0.4)
plt.show()
```

```python
plt.scatter(x, y, label='Samples')
plt.legend()
plt.show()
```

```python
plt.plot(x, y, 'o')
```
